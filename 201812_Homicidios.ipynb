```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('fivethirtyeight')

%matplotlib inline
```

```python
# Leer datos de poblacion y de homicidios entre 2010 y noviembre 2017
homi = pd.read_csv('201812_homi.csv',encoding = 'iso-8859-1',delimiter = ';')
pobla = pd.read_csv('201812_Poblacion.csv.txt',encoding = 'iso-8859-1',delimiter = ';')

pobla.drop(pobla.columns[9], axis=1, inplace = True)
homi.drop(homi.columns[9], axis=1, inplace = True)
```

```python
# Unir ambos ficheros y calcular homicidios por cada 100.000 habitantes para cada año de la serie
result = pd.merge(pobla, homi, on='Provincia')

result['r2010']=result['h2010']/result['p2010']*100000
result['r2011']=result['h2011']/result['p2011']*100000
result['r2012']=result['h2012']/result['p2012']*100000
result['r2013']=result['h2013']/result['p2013']*100000
result['r2014']=result['h2014']/result['p2014']*100000
result['r2015']=result['h2015']/result['p2015']*100000
result['r2016']=result['h2016']/result['p2016']*100000
result['r2017']=result['h2017']/result['p2017']*100000

result['media']=(result['h2010']+result['h2011']+result['h2012']+result['h2013']+result['h2014']+result['h2015']+result['h2016']+result['h2017'])/(result['p2010']+result['p2011']+result['p2012']+result['p2013']+result['p2014']+result['p2015']+result['p2016']+result['p2017'])*100000

result[['Provincia','r2017','r2016','r2015','r2014','r2013','r2012','r2011','r2010', 'media']]
```

```python
# Ver resultados de Total España (línea 0) y Madrid (línea 28)
print(result.ix[0,"r2017":"r2010"])
print(result.ix[28,"r2017":"r2010"])
```

```python
x = np.arange(8)
labels = ['2010','2011','2012','2013','2014','2015','2016','2017']

# Ratios de España(0) y de Madrid (28)
y1 = result.ix[0,"r2010":"r2017"]
y2 = result.ix[28,"r2010":"r2017"]


params = {'legend.fontsize': 14,
          'figure.figsize': (10, 5),
         'axes.labelsize': 14,
         'axes.titlesize':14,
         'xtick.labelsize':14,
         'ytick.labelsize':14}
plt.rcParams.update(params)

plt.plot(x,y1, label='España', color=sns.color_palette()[3],ls='--', alpha=0.5)
plt.plot(x,y2, label='Madrid', color=sns.color_palette()[0])

plt.xticks(x, labels, rotation='horizontal')
plt.axis(ymin= 0 , ymax=1.00)
plt.title('Homicidios por cada 100.000 habitantes',fontsize= 18)
plt.legend()
plt.savefig('images/homicidios.png', bbox_inches='tight', pad_inches=0.3, dpi=220)
plt.show()
```
