```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('fivethirtyeight')

%matplotlib inline
```

```python
# Leer datos demogr√°ficos de enero 2016 y noviembre 2018
actuac = pd.read_csv('201812_Actuaciones.csv',encoding = 'iso-8859-1',delimiter = ';')

actuac
```

```python
x = actuac.index
labels = actuac['year']

# Actuaciones
y1 = actuac['act_seg']
y2 = actuac['venta_amb']
y3 = actuac['Inspecc_locales']
y4 = actuac['patrullas']



params = {'legend.fontsize': 14,
          'figure.figsize': (10, 10),
         'axes.labelsize': 14,
         'axes.titlesize':14,
         'xtick.labelsize':14,
         'ytick.labelsize':14}
plt.rcParams.update(params)

plt.plot(x,y1, label='Actuaciones de Seguridad', color=sns.color_palette()[3])
plt.plot(x,y2, label='Venta Ambulante', color=sns.color_palette()[0])
plt.plot(x,y3, label='Inspecciones de Locales', color=sns.color_palette()[1])
plt.plot(x,y4, label='Patrullas semanales', color=sns.color_palette()[2])

plt.xticks(x, labels, rotation='horizontal')
plt.axis(ymin= 0 , ymax=7000)
plt.title('Actuaciones de la Policia Municipal de Madrid',fontsize= 18)
plt.legend(loc='right')
plt.savefig('images/actuaciones.png', bbox_inches='tight', pad_inches=0.3, dpi=220)
plt.show()
```
